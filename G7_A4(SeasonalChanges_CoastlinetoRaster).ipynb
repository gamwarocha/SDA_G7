```python
#Chla conc. North Sea in January 2025 (Winter)
import xarray as xr
import matplotlib.pyplot as plt
import geopandas as gpd
import cartopy.crs as ccrs
from matplotlib_scalebar.scalebar import ScaleBar

# Load CHLA data
data = xr.open_dataset("./chla_2025/mercatorbiomer4v2r1_global_mean_pft_202501.nc")

# Select surface chlorophyll-a
surface_chl = data["chl"].isel(time=0, depth=0)

# Clip to North Sea
north_sea_chl = surface_chl.sel(
    longitude=slice(-4, 9),
    latitude=slice(52, 62)
)

# Load coastline
coastline = gpd.read_file("./ne_10m_coastline/ne_10m_coastline.shp")
coastline = coastline.to_crs("EPSG:4326")

# Use km/metre-based projection for Europe
proj = ccrs.epsg(3035)

fig = plt.figure(figsize=(10, 7))
ax = plt.axes(projection=proj)

# Set map extent in lon/lat
ax.set_extent([-4, 9, 52, 62], crs=ccrs.PlateCarree())

# Plot CHLA raster
north_sea_chl.plot(
    ax=ax,
    transform=ccrs.PlateCarree(),
    cmap="viridis",
    cbar_kwargs={"label": "Chlorophyll-a concentration (mg/m³)"}
)

# Plot coastline
ax.add_geometries(
    coastline.geometry,
    crs=ccrs.PlateCarree(),
    facecolor="none",
    edgecolor="black",
    linewidth=0.8
)

# Gridlines
gl = ax.gridlines(draw_labels=True, linewidth=0.3, alpha=0.5)
gl.top_labels = False
gl.right_labels = False

# North arrow
ax.annotate(
    "N",
    xy=(0.92, 0.88),
    xytext=(0.92, 0.78),
    arrowprops=dict(facecolor="black", width=4, headwidth=12),
    ha="center",
    va="center",
    fontsize=12,
    xycoords=ax.transAxes
)

# Scale bar
scalebar = ScaleBar(
    dx=1,
    units="m",
    location="lower right",
    scale_loc="bottom"
)
ax.add_artist(scalebar)

# CRS note
ax.text(
    0.01, 0.02,
    "Projected CRS: ETRS89 / LAEA Europe (EPSG:3035)",
    transform=ax.transAxes,
    fontsize=9,
    bbox=dict(facecolor="white", alpha=0.8, edgecolor="none")
)

ax.set_title("Surface Chlorophyll-a Concentration in the North Sea (January 2025) - Winter")

plt.show()
```

**Figure 1a.** *Spatial distribution of surface chlorophyll-a concentration in the North Sea during January 2025.*

Higher chlorophyll-a concentrations were mainly found near coastal regions, likely due to higher nutrient inputs from rivers and coastal mixing, which support phytoplankton growth. 

In contrast, offshore areas showed lower chlorophyll-a levels, possibly due to lower nutrient availability and more stable ocean conditions. 

These patterns are consistent with previous studies linking phytoplankton distribution to nutrient availability and environmental conditions (Boyer et al., 2009).

This pattern may show differences in nutrient levels, ocean currents, and environmental conditions across the study area.

### 4.2 Comparison of Seasonal Surface Chlorophyll-a Concentration Distributions in the North Sea 2025

```python
import xarray as xr
import matplotlib.pyplot as plt

# Open datasets
jan = xr.open_dataset("./chla_2025/mercatorbiomer4v2r1_global_mean_pft_202501.nc")
apr = xr.open_dataset("./chla_2025/mercatorbiomer4v2r1_global_mean_pft_202504.nc")
jul = xr.open_dataset("./chla_2025/mercatorbiomer4v2r1_global_mean_pft_202507.nc")
octo = xr.open_dataset("./chla_2025/mercatorbiomer4v2r1_global_mean_pft_202510.nc")

# Function to extract surface CHLA and clip to North Sea
def get_north_sea_chl(ds):
    chl = ds["chl"].isel(time=0, depth=0)
    chl_ns = chl.sel(
        longitude=slice(-4, 9),
        latitude=slice(52, 62)
    )
    return chl_ns

jan_ns = get_north_sea_chl(jan)
apr_ns = get_north_sea_chl(apr)
jul_ns = get_north_sea_chl(jul)
oct_ns = get_north_sea_chl(octo)

# Plot histogram comparison
plt.figure(figsize=(10, 6))

jan_ns.plot.hist(bins=50, alpha=0.45, label="January 2025 (Winter)")
apr_ns.plot.hist(bins=50, alpha=0.45, label="April 2025 (Spring)")
jul_ns.plot.hist(bins=50, alpha=0.45, label="July 2025 (Summer)")
oct_ns.plot.hist(bins=50, alpha=0.45, label="October 2025 (Autumn)")

plt.title("Comparison of Seasonal Surface Chlorophyll-a Concentration Distributions")
plt.xlabel("Chlorophyll-a concentration (mg/m³)")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

## 4.3 Monthly Mean CHLA Trend

```python
import glob
import xarray as xr
import matplotlib.pyplot as plt

files = sorted(glob.glob("./chla_2025/*.nc"))

#print("Number of files:", len(files))

monthly_mean = []

for file in files:
    ds = xr.open_dataset(file)

    chl = ds["chl"].isel(time=0, depth=0)

    north_sea = chl.sel(
        longitude=slice(-4, 9),
        latitude=slice(52, 62)
    )

    monthly_mean.append(float(north_sea.mean()))

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", 
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

plt.figure(figsize=(10,6))

plt.plot(months, monthly_mean, marker="o")

plt.title("Monthly Mean Surface Chlorophyll-a Concentration in the North Sea (2025)")
plt.xlabel("Month")
plt.ylabel("Mean Chlorophyll-a (mg/m³)")
plt.grid(True)

plt.show()
```

**Figure 3.** *Monthly Mean Surface Chlorophyll-a Concentration in the North Sea during 2025*

```python
# Annual Mean CHLA 2025 + Oil Wells Map

# Uncomment if required
# !pip install xarray netCDF4 geopandas cartopy matplotlib-scalebar

import glob
import xarray as xr
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
from matplotlib_scalebar.scalebar import ScaleBar

# -----------------------------
# 1. Load all CHLA files
# -----------------------------

files = sorted(glob.glob("./chla_2025/*.nc"))

#print("Number of CHLA files:", len(files))

monthly_chl = []

for file in files:
    ds = xr.open_dataset(file)

    chl = ds["chl"].isel(time=0, depth=0)

    north_sea = chl.sel(
        longitude=slice(-4, 9),
        latitude=slice(52, 62)
    )

    monthly_chl.append(north_sea)

# Calculate annual mean CHLA
annual_mean_chl = xr.concat(monthly_chl, dim="month").mean(dim="month")

# -----------------------------
# 2. Load oil wells
# -----------------------------

oil_df = pd.read_csv(
    "./UKCS_offshore_petroleum_exploration_and_appraisal_well_results_(ED50)_-1618047620638058084.csv"
)

oil_wells = gpd.GeoDataFrame(
    oil_df,
    geometry=gpd.points_from_xy(oil_df["x"], oil_df["y"]),
    crs="EPSG:4326"
)

oil_wells_ns = oil_wells.cx[-4:9, 52:62]

# -----------------------------
# 3. Load coastline
# -----------------------------

coastline = gpd.read_file("./ne_10m_coastline/ne_10m_coastline.shp")
coastline = coastline.to_crs("EPSG:4326")

# -----------------------------
# 4. Plot with projected CRS
# -----------------------------

projected_crs = ccrs.epsg(3035)        # ETRS89 / LAEA Europe
geographic_crs = ccrs.PlateCarree()    # lon/lat source CRS

fig = plt.figure(figsize=(11, 8))
ax = plt.axes(projection=projected_crs)

# Set North Sea extent
ax.set_extent([-4, 9, 52, 62], crs=geographic_crs)

# Plot annual mean CHLA raster
annual_mean_chl.plot(
    ax=ax,
    transform=geographic_crs,
    cmap="viridis",
    cbar_kwargs={
        "label": "Annual mean chlorophyll-a concentration (mg/m³)",
        "shrink": 0.75
    }
)

# Plot coastline
ax.add_geometries(
    coastline.geometry,
    crs=geographic_crs,
    facecolor="none",
    edgecolor="black",
    linewidth=0.8
)

# Plot oil wells
ax.scatter(
    oil_wells_ns.geometry.x,
    oil_wells_ns.geometry.y,
    transform=geographic_crs,
    color="red",
    s=12,
    label="Oil wells",
    edgecolor="black",
    linewidth=0.2
)

# Gridlines
gl = ax.gridlines(
    draw_labels=True,
    linewidth=0.3,
    alpha=0.5
)
gl.top_labels = False
gl.right_labels = False

# North arrow
ax.annotate(
    "N",
    xy=(0.93, 0.88),
    xytext=(0.93, 0.76),
    arrowprops=dict(facecolor="black", width=4, headwidth=12),
    ha="center",
    va="center",
    fontsize=13,
    fontweight="bold",
    xycoords=ax.transAxes
)

# Scale bar
scalebar = ScaleBar(
    dx=1,
    units="m",
    location="lower right",
    scale_loc="bottom",
    length_fraction=0.25
)
ax.add_artist(scalebar)

# CRS note
ax.text(
    0.01,
    0.02,
    "Projected CRS: ETRS89 / LAEA Europe (EPSG:3035)",
    transform=ax.transAxes,
    fontsize=9,
    bbox=dict(facecolor="white", alpha=0.8, edgecolor="none")
)

# Title and legend
ax.set_title(
    "Annual Mean Surface Chlorophyll-a Concentration and Oil Wells in the North Sea (2025)",
    fontsize=13
)

ax.legend(loc="upper left")

plt.show()
```

```python
# Coastline to Raster Conversion + Distance-to-Coast Raster

# Uncomment if required
# !pip install geopandas rasterio scipy matplotlib

import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
import rasterio
from rasterio.features import rasterize
from rasterio.transform import from_bounds
from shapely.geometry import box
from scipy.ndimage import distance_transform_edt

# -----------------------------
# 1. Load coastline
# -----------------------------

coastline = gpd.read_file("./ne_10m_coastline/ne_10m_coastline.shp")
coastline = coastline.to_crs("EPSG:4326")

# -----------------------------
# 2. Create North Sea study area
# -----------------------------

north_sea_box = gpd.GeoDataFrame(
    geometry=[box(-4, 52, 9, 62)],
    crs="EPSG:4326"
)

# Clip coastline to North Sea area
north_sea_coast = gpd.clip(coastline, north_sea_box)

# -----------------------------
# 3. Reproject to metre-based CRS
# -----------------------------

north_sea_coast_m = north_sea_coast.to_crs("EPSG:3035")
north_sea_box_m = north_sea_box.to_crs("EPSG:3035")

# -----------------------------
# 4. Create raster grid
# -----------------------------

xmin, ymin, xmax, ymax = north_sea_box_m.total_bounds

# Raster resolution in metres
# 10000 = 10 km grid cells
resolution = 10000

width = int((xmax - xmin) / resolution)
height = int((ymax - ymin) / resolution)

transform = from_bounds(
    xmin,
    ymin,
    xmax,
    ymax,
    width,
    height
)

# -----------------------------
# 5. Rasterize coastline
# -----------------------------

coastline_raster = rasterize(
    [(geom, 1) for geom in north_sea_coast_m.geometry],
    out_shape=(height, width),
    transform=transform,
    fill=0,
    dtype="uint8"
)

# -----------------------------
# 6. Create distance-to-coast raster
# -----------------------------

# distance_transform_edt calculates distance in pixels
# Multiply by resolution to convert pixels to metres
distance_to_coast_m = distance_transform_edt(coastline_raster == 0) * resolution

# Convert metres to kilometres
distance_to_coast_km = distance_to_coast_m / 1000

# -----------------------------
# 7. Plot rasterized coastline
# -----------------------------

plt.figure(figsize=(8, 8))
plt.imshow(coastline_raster, cmap="gray")
plt.title("Rasterized Coastline within North Sea Study Area")
plt.axis("off")
plt.show()

# -----------------------------
# 8. Plot distance-to-coast raster
# -----------------------------

plt.figure(figsize=(10, 8))
plt.imshow(distance_to_coast_km, cmap="viridis")
plt.colorbar(label="Distance to coastline (km)")
plt.title("Distance-to-Coast Raster for the North Sea")
plt.axis("off")
plt.show()

# -----------------------------
# 9. Save distance raster as GeoTIFF
# -----------------------------

output_path = "./distance_to_coast_km.tif"

with rasterio.open(
    output_path,
    "w",
    driver="GTiff",
    height=distance_to_coast_km.shape[0],
    width=distance_to_coast_km.shape[1],
    count=1,
    dtype=distance_to_coast_km.dtype,
    crs="EPSG:3035",
    transform=transform
) as dst:
    dst.write(distance_to_coast_km, 1)

print("Distance-to-coast raster saved as:", output_path)
```
